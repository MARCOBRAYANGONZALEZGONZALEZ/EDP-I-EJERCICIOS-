<font color = 'blue'>Grafica de la parabola: $y=x^2$.<font>


Este codigo grafica una parabola dada una funcion, utilizando bibliotecas con contenido matematico.


```python
# prompt: dame un codigo que grafique y= x al cuadrado

import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.linspace(-10, 10, 400)

# Calculate y values
y = x**2

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.title("Grafica de y = x^2")

# Display the plot
plt.grid(True)
plt.show()
```
