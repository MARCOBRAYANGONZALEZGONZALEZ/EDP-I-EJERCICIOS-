```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parámetros de la malla
Nx = 51  # puntos en x
Ny = 51  # puntos en y
hx = 1.0 / (Nx - 1)
hy = 1.0 / (Ny - 1)

# Crear malla
x = np.linspace(0, 1, Nx)
y = np.linspace(0, 1, Ny)
X, Y = np.meshgrid(x, y)

# Inicializar solución
U = np.zeros((Ny, Nx))

# Aplicar condiciones de Dirichlet
# Condición inferior: y = 0
U[0, :] = 0
# Condición izquierda: x = 0
U[:, 0] = 0
# Condición derecha: x = 1
U[:, -1] = y  # u(1,y) = y

# Inicializar la fila superior (y=1) usando condición de Neumann
# Primera aproximación: u(x,1) = u(x,1-hy) + hy * x^2
for i in range(Nx):
    U[-1, i] = U[-2, i] + hy * x[i]**2

# Método iterativo de Gauss-Seidel para ecuación de Poisson
max_iter = 20000
tolerance = 1e-8

print("Resolviendo ecuación de Poisson con condición de Neumann...")

for iteration in range(max_iter):
    U_old = U.copy()

    # Actualizar puntos interiores
    for i in range(1, Nx-1):
        for j in range(1, Ny-1):
            # Término fuente f(x,y) = 2y
            f = 2 * y[j]

            # Coeficientes
            coeff_x = 1 / hx**2
            coeff_y = 1 / hy**2
            coeff_total = 2 * (coeff_x + coeff_y)

            U[j, i] = (coeff_x * (U[j, i+1] + U[j, i-1]) +
                       coeff_y * (U[j+1, i] + U[j-1, i]) - f) / coeff_total

    # Actualizar frontera superior (condición de Neumann)
    for i in range(1, Nx-1):
        # Usamos aproximación de segundo orden para condición de Neumann
        # (u_{i,Ny-1} - u_{i,Ny-3})/(2hy) = x_i^2
        # Esto da: u_{i,Ny-1} = u_{i,Ny-3} + 2hy * x_i^2
        U[-1, i] = U[-3, i] + 2 * hy * x[i]**2

    # Calcular cambio máximo
    max_change = np.max(np.abs(U - U_old))

    if iteration % 1000 == 0:
        print(f"Iteración {iteration}: cambio máximo = {max_change:.2e}")

    if max_change < tolerance:
        print(f"\nConvergencia alcanzada en {iteration+1} iteraciones")
        break

# Solución exacta
U_exact = X**2 * Y

# Error numérico
error = np.abs(U - U_exact)
max_error = np.max(error)
mean_error = np.mean(error)
rms_error = np.sqrt(np.mean(error**2))

print(f"\nError máximo: {max_error:.6e}")
print(f"Error medio: {mean_error:.6e}")
print(f"Error RMS: {rms_error:.6e}")

# Verificación de condición de Neumann
print("\nVerificación condición de Neumann en y=1:")
neumann_numeric = (U[-1, :] - U[-2, :]) / hy
neumann_exact = x**2
neumann_error = np.max(np.abs(neumann_numeric - neumann_exact))
print(f"Error en condición de Neumann: {neumann_error:.6e}")

# Paso 4: Visualización
fig = plt.figure(figsize=(18, 6))

# Gráfico 1: Solución numérica
ax1 = fig.add_subplot(131, projection='3d')
surf1 = ax1.plot_surface(X, Y, U, cmap='viridis', alpha=0.8)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('u(x,y)')
ax1.set_title('Solución Numérica de Poisson')
ax1.view_init(elev=30, azim=45)
fig.colorbar(surf1, ax=ax1, shrink=0.5)

# Gráfico 2: Solución exacta
ax2 = fig.add_subplot(132, projection='3d')
surf2 = ax2.plot_surface(X, Y, U_exact, cmap='plasma', alpha=0.8)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('u(x,y)')
ax2.set_title('Solución Exacta: $u(x,y)=x^2y$')
ax2.view_init(elev=30, azim=45)
fig.colorbar(surf2, ax=ax2, shrink=0.5)

# Gráfico 3: Error
ax3 = fig.add_subplot(133, projection='3d')
surf3 = ax3.plot_surface(X, Y, error, cmap='hot', alpha=0.8)
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_zlabel('Error')
ax3.set_title(f'Error (máx: {max_error:.2e})')
ax3.view_init(elev=30, azim=45)
fig.colorbar(surf3, ax=ax3, shrink=0.5)

plt.tight_layout()
plt.show()

# Gráfico 4: Comparación en secciones transversales
fig2, axes = plt.subplots(2, 3, figsize=(15, 10))

# Sección en y = 0.5
y_idx = np.argmin(np.abs(y - 0.5))
axes[0, 0].plot(x, U[y_idx, :], 'b-', linewidth=2, label='Numérica')
axes[0, 0].plot(x, U_exact[y_idx, :], 'r--', linewidth=2, label='Exacta')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('u(x, 0.5)')
axes[0, 0].set_title('Sección en y = 0.5')
axes[0, 0].legend()
axes[0, 0].grid(True)

# Sección en x = 0.5
x_idx = np.argmin(np.abs(x - 0.5))
axes[0, 1].plot(y, U[:, x_idx], 'b-', linewidth=2, label='Numérica')
axes[0, 1].plot(y, U_exact[:, x_idx], 'r--', linewidth=2, label='Exacta')
axes[0, 1].set_xlabel('y')
axes[0, 1].set_ylabel('u(0.5, y)')
axes[0, 1].set_title('Sección en x = 0.5')
axes[0, 1].legend()
axes[0, 1].grid(True)

# Condición de Neumann en y=1
axes[0, 2].plot(x, neumann_numeric, 'b-', linewidth=2, label='Numérica')
axes[0, 2].plot(x, neumann_exact, 'r--', linewidth=2, label='Exacta')
axes[0, 2].set_xlabel('x')
axes[0, 2].set_ylabel('∂u/∂y(x, 1)')
axes[0, 2].set_title('Condición de Neumann en y=1')
axes[0, 2].legend()
axes[0, 2].grid(True)

# Error en y = 0.5
axes[1, 0].plot(x, error[y_idx, :], 'g-', linewidth=2)
axes[1, 0].set_xlabel('x')
axes[1, 0].set_ylabel('Error')
axes[1, 0].set_title(f'Error en y = 0.5')
axes[1, 0].grid(True)
axes[1, 0].set_ylim([0, np.max(error[y_idx, :]) * 1.1])

# Error en x = 0.5
axes[1, 1].plot(y, error[:, x_idx], 'g-', linewidth=2)
axes[1, 1].set_xlabel('y')
axes[1, 1].set_ylabel('Error')
axes[1, 1].set_title(f'Error en x = 0.5')
axes[1, 1].grid(True)
axes[1, 1].set_ylim([0, np.max(error[:, x_idx]) * 1.1])

# Error en condición de Neumann
axes[1, 2].plot(x, np.abs(neumann_numeric - neumann_exact), 'm-', linewidth=2)
axes[1, 2].set_xlabel('x')
axes[1, 2].set_ylabel('Error')
axes[1, 2].set_title(f'Error en condición de Neumann')
axes[1, 2].grid(True)

plt.tight_layout()
plt.show()

# Gráfico 5: Mapa de contorno del error
fig3, ax = plt.subplots(figsize=(10, 8))
contour = ax.contourf(X, Y, error, levels=50, cmap='hot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Mapa de contorno del error absoluto')
fig3.colorbar(contour, ax=ax)
ax.set_aspect('equal')
plt.show()
```
