```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
c = 1  # velocidad de propagación
t = 0.5  # tiempo
x = np.linspace(0, 1, 500)  # dominio espacial

# Definición de la función f(x')
def f(x_):
    return np.exp(-((x_ - 0.5)**2) / 0.01)

# Inciso (a): x' = x - ct -> onda viajera a la derecha
f_derecha = f(x - c * t)

# Inciso (b): x' = x + ct -> onda viajera a la izquierda
f_izquierda = f(x + c * t)

# Inciso (c): superposición de ambas ondas
f_superpuesta = f_derecha + f_izquierda

# Graficar
plt.figure(figsize=(12, 8))

# Onda viajera a la derecha
plt.subplot(3, 1, 1)
plt.plot(x, f_derecha, label=r'$f(x - ct)$', color='blue')
plt.title('Onda viajera a la derecha')
plt.grid(True)
plt.legend()

# Onda viajera a la izquierda
plt.subplot(3, 1, 2)
plt.plot(x, f_izquierda, label=r'$f(x + ct)$', color='green')
plt.title('Onda viajera a la izquierda')
plt.grid(True)
plt.legend()

# Superposición
plt.subplot(3, 1, 3)
plt.plot(x, f_superpuesta, label=r'$f(x - ct) + f(x + ct)$', color='red')
plt.title('Superposición de ondas viajeras')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
